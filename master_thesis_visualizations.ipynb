```python
import os
import json

dataset = 'taxi'
layer = 'StartLocationLayer'

with open(f'data/{dataset}/raw/clusters/{layer}.json', 'r') as fh:
    clusters = json.loads(fh.read())
    
```

```python
# remove noise cluster
clusters[0]['cluster_label']
clusters = clusters[1:]
```

```python
class Info:
    def __init__(self, size, center):
        self.size = size
        self.center = center
```

```python
from processing import ClusterMetricsCalculator2D

infos = []
for c in clusters:
    calc = ClusterMetricsCalculator2D(c['nodes'], ['start_location_lat', 'start_location_long'], 0, 0)   
    infos.append(Info(calc.get_size(), calc.get_center()))
```

```python
points = [i for i in infos if i.size > 70]
len(points)
```

```python
import matplotlib.pyplot as plt

plt.scatter([p.center[0] for p in points],
            [p.center[1] for p in points],
            s=[p.size/50 for p in points]
            )

plt.show()
```

```python
from typing import List
import folium

def displayNodes(nodes: List[List[float]]):  
    '''
    Displays the nodes on a map of the city.

    :param nodes: A list of coordinates, eg. [[1,2],[1,3]]
    '''
    m = folium.Map(location=[41.15,-8.6],tiles='stamenterrain',zoom_start=12, control_scale=True) 
    # marker_cluster = folium.MarkerCluster().add_to(m)

    for idx, node in enumerate(nodes): 
        popupLabel = idx

        folium.Marker(
          location=[node[1], node[0]],
          #popup='Cluster Nr: '+ str(node.cluster_no),
          popup=popupLabel,
          icon=folium.Icon(color='red', icon='circle'),
        ).add_to(m)
      
    display(m)
```

```python
displayNodes([p.center for p in points])
```
